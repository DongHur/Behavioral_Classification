```python
import pandas as pd
import numpy as np
from numpy import linalg as LA
from sklearn import cluster

import matplotlib
import matplotlib.pyplot as plt

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go

from tqdm import tqdm
```

```python
# Import Data
data_path = "test_data/zoomed_view_editDeepCut_resnet50_antennatrackingJun19shuffle1_1030000.h5"
df = pd.read_hdf(data_path)['DeepCut_resnet50_antennatrackingJun19shuffle1_1030000']
df_data = df.as_matrix()
df
```

```python
# REFORMAT DATA
data={}
# NUMBER OF BODYPARTS
num_time_points = len(df_data)
num_col = len(df_data[0])
num_features = num_col/3 # 30

text = []
for i in range(int(num_col/3)):
    data['bp'+str(i)] = np.array(df_data[:,i*3:i*3+3])
    text.append(str(i))
```

```python
coord = np.zeros((30,2))
frame = 1052
text = []
for i in range(30):
    coord[i,:] = data['bp'+str(i)][frame,0:2]
    text.append(str(i))
    
# GRAPH
data_graph = [go.Scatter(
    x=coord[:,0], y=coord[:,1], name='Frame', marker=dict(color='rgb(255,0,0)'), mode = 'markers',
    text=text)
 ]
layout = go.Layout(
  title = 'Coordinates for 1 Frame',
  xaxis=dict(title="X"),
  yaxis=dict(title="Y")
)
fig1 = go.Figure(data=data_graph, layout=layout)
plot(fig1)
```
