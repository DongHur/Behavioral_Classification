```python
import pandas as pd
import numpy as np
from numpy import linalg as LA
from sklearn import cluster

import matplotlib
import matplotlib.pyplot as plt

from tqdm import tqdm
```

```python
# Import Data
data_path = "ant_dotsDeepCut_resnet50_antennatrackingJun19shuffle1_1030000.h5"
df = pd.read_hdf(data_path)['DeepCut_resnet50_antennatrackingJun19shuffle1_1030000']
df_data = df.as_matrix()
```

```python
# REFORMAT DATA
data={}
text = []
# NUMBER OF BODYPARTS
num_time_points = len(df_data)
num_col = len(df_data[0])
num_features = num_col/3 # 30
# POPULATE DATA
for i in range(int(num_col/3)):
    data['bp'+str(i)] = np.array(df_data[:,i*3:i*3+3])
    text.append(str(i))
```

```python
print(data['bp0'].shape)
# X ==> data['bp0'][:,0]
# Y ==> data['bp0'][:,1]
# L ==> data['bp0'][:,2]
```

### K-Means Clustering

```python
# GRAPH DATA
bp_label = 'bp0'
bp0x = data[bp_label][:,0]
bp0y = data[bp_label][:,1]
plt.scatter(bp0x,bp0y)
plt.title('Body Part 0 - X and Y position')
plt.xlabel('Body Part 0 - X')
plt.ylabel('Body Part 0 - Y')
plt.show()
```

```python
# SAMPLE TEST DATA
from sklearn.datasets import make_blobs
n_samples = 10000
random_state = 100
X, clusters = make_blobs(n_samples=n_samples, random_state=random_state)
# DATA
print(X.shape[0])
print(X.shape[1])

k = 5
k_means = cluster.KMeans(n_clusters=3, init='random', n_init=100, random_state=0).fit(X)
print("Labels: ", k_means.labels_)
print("Clusteroids: ", k_means.cluster_centers_)

x = X[:,0]
y = X[:,1]
x_clusteroid = k_means.cluster_centers_[:,0]
y_clusteroid = k_means.cluster_centers_[:,1]

# GRAPH DATA
plt.scatter(x, y, c=k_means.labels_, marker='.')
plt.scatter(x_clusteroid, y_clusteroid, s=100, marker='*')
plt.title('X and Y position')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# ACTUAL DATA
# CONSTANTS INSTANTIATION
body_part = 'bp2'
k = 4
epoch = 100

# DATA
bp0x = np.transpose(data[body_part][:,0])
bp0y = np.transpose(data[body_part][:,1])
X = np.stack((bp0x,bp0y),axis=-1)
print(X.shape[0])
print(X.shape[1])

k_means = cluster.KMeans(n_clusters=k, init='random', n_init=epoch, random_state=0).fit(X)
print("Body Part: ", body_part)
print("Labels: ", k_means.labels_)
print("Clusteroids: ", k_means.cluster_centers_)
print("Intertia: ", k_means.inertia_)

# GRAPH DATA
x = X[:,0]
y = X[:,1]
x_clusteroid = k_means.cluster_centers_[:,0]
y_clusteroid = k_means.cluster_centers_[:,1]
plt.scatter(x, y, c=k_means.labels_, marker='.')
plt.scatter(x_clusteroid, y_clusteroid, s=100, marker='*')
plt.title('X and Y position')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# ONE BODYPART X AND Y COORDINATE CLUSTERING
# ELBOW PLOT
# CONSTANT INSTANTIATION
body_part = 'bp2'
min_k = 3
max_k = 11
epoch = 1000
rand_seed = 0

#******************#
# DATA
bp_x = np.transpose(data[body_part][:,0])
bp_y = np.transpose(data[body_part][:,1])
X = np.stack((bp_x,bp_y),axis=-1)


k_array = np.arange(min_k,max_k)
sse = np.zeros(k_array.shape[0])

for k_i, k in tqdm(enumerate(k_array)):
    k_means = cluster.KMeans(n_clusters=k, init='random', n_init=epoch, \
                             random_state=rand_seed).fit(X)
    sse[k_i] = k_means.inertia_
    
# GRAPH ELBOW
plt.plot(k_array, sse)
plt.title('Elbow Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors (Inertia)')
plt.show()

# DATA INFO
print("Number of Data Points: ", X.shape[0])
print("Coordinate: ", X.shape[1])
    
```

```python
# ALL BODYPART X AND Y COORDINATE CLUSTERING
# CONSTANT INSTANTIATION
min_k = 3
max_k = 11
epoch = 1000
rand_seed = 0

#******************#
# DATA
bp_x = np.transpose(data['bp0'][:,0])
bp_y = np.transpose(data['bp0'][:,1])
X = np.stack((bp_x,bp_y),axis=-1)

for i in tqdm(range(1,30)):
    bp_x = np.transpose(data["bp"+str(i)][:,0])
    bp_y = np.transpose(data["bp"+str(i)][:,1])
    X_temp = np.stack((bp_x,bp_y),axis=-1)
    X = np.append(X, X_temp, -1)


k_array = np.arange(min_k,max_k)
sse = np.zeros(k_array.shape[0])

for k_i, k in tqdm(enumerate(k_array)):
    k_means = cluster.KMeans(n_clusters=k, init='random', n_init=epoch, \
                             random_state=rand_seed).fit(X)
    sse[k_i] = k_means.inertia_
    
# GRAPH ELBOW
plt.plot(k_array, sse)
plt.title('Elbow Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors (Inertia)')
plt.show()

# DATA INFO
print("Number of Data Points: ", X.shape[0])
print("Coordinate: ", X.shape[1])
```
