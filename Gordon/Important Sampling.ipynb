```python
import numpy as np
import torch
import matplotlib
import matplotlib.pyplot as plt
from tqdm import tqdm

from tsne_python.tsne import tsne
from Sampling.Importance_Sampling import Sampling
from tsne_python.tsne_embed import tsne_embed
from tqdm import tqdm, tqdm_notebook
```

```python
power_spec = torch.Tensor(np.load("CB_power_100data_60bp_25fBin.npy"))
(num_bp, num_coord, num_freq, num_frame) = power_spec.shape
pwr = torch.reshape(power_spec, (num_bp*num_coord*num_freq, num_frame)).t()
amp = pwr.sum(dim=1)
pwr = (pwr.t()/amp).t().numpy()
# RUN tSNE
pwr_p = tsne(X=pwr, no_dims=2, initial_dims=50, perplexity=20.0, max_iter=1000)
pwr_p = torch.Tensor(pwr_p)
```

```python
k_nearest = 5
num_points = 50
# y_pwr and pwr should be randomly selected points from the total data!!!!
# DONT RUN ON EVERYTHING. OR ELSE IT WILL TAKE TOO LONG
Sample = Sampling(pwr_p, pwr, k_nearest=k_nearest, num_points=num_points)
K_idx, K_matrix_idx, K_dist = Sample.knn()
GH_conv=Sample.gaussian_conv(K_dist)
watershed_dist, labels = Sample.apply_watershed(GH_conv, min_distance=6)
Template = Sample.create_template(labels)
training_data = Sample.importance_sampling(Template, frac_training_set=0.95)
```

```python
# RUN tSNE ON TRAINIG SET AFTER SUBSAMPLING
X = training_data
Xp= tsne(X=X.numpy(), no_dims=2, initial_dims=50, perplexity=20.0, max_iter=1000)
Xp= torch.Tensor(Xp)
```

```python
Y = torch.Tensor(pwr)
Yp, P, Q = tsne_embed(X=X, Xp=Xp, Y=Y, no_dims=2, initial_dims=50, perplexity=30.0, max_iter=1000, plot=False)
```

```python
plt.scatter(Xp[:,0].numpy(), Xp[:,1].numpy())
plt.scatter(Yp[:,0].numpy(), Yp[:,1].numpy())
plt.title("Training + Emedded Points")
plt.xlabel("X1")
plt.ylabel("X2")
plt.axis('scaled')
plt.show()
```

```python
Tpwr = torch.cat((X, Y), dim=0)
Tpwr_p = torch.cat((Xp, Yp), dim=0)
k_nearest = 4
num_points = 50
# CREATE DENSITY
Sample = Sampling(Tpwr_p, Tpwr, k_nearest=k_nearest, num_points=num_points)
K_idx, K_matrix_idx, K_dist = Sample.knn()
GH_conv=Sample.gaussian_conv(K_dist)
watershed_dist, labels = Sample.apply_watershed(GH_conv, min_distance=6)
```
